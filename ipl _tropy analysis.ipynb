```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import seaborn as sns 
 



data = {
    'Year': range(2008, 2025),
    'Winner': [
        'Rajasthan Royals', 'Deccan Chargers', 'Chennai Super Kings', 'Chennai Super Kings',
        'Kolkata Knight Riders', 'Mumbai Indians', 'Kolkata Knight Riders', 'Mumbai Indians',
        'Sunrisers Hyderabad', 'Mumbai Indians', 'Chennai Super Kings', 'Mumbai Indians',
        'Mumbai Indians', 'Chennai Super Kings', 'Gujarat Titans', 'Chennai Super Kings', 'Kolkata Knight Riders'
    ]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Count the number of trophies won by each team
trophy_counts = df['Winner'].value_counts()
colors = ['yellow', 'blue', 'purple', 'pink', 'violet', 'orange', 'indigo']
# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(x=trophy_counts.index, y=trophy_counts.values, palette='viridis')
plt.title('Number of IPL Trophies Won by Each Team (2008-2024)')
plt.xlabel('Team')
plt.ylabel('Number of Trophies')
plt.xticks(rotation=45)
legend_elements = [Patch(facecolor=color, label=label) for color, label in zip(colors, ['CSK', 'MI', 'KKR', 'RR', 'DC', 'SRH', 'GT'])]
plt.legend(handles=legend_elements, title="Teams")
plt.show()
```

```python
# Plot the distribution of wins over the years
plt.figure(figsize=(12, 6))
sns.countplot(x='Winner', hue='Year', data=df, palette='Set2', dodge=False)
plt.title('Distribution of IPL Wins Over the Years (2008-2024)')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Create a pivot table for the heatmap
pivot_table = df.pivot_table(index='Winner', columns='Year', aggfunc='size', fill_value=0)

# Plot the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='d')
plt.title('IPL Wins by Team and Year (2008-2024)')
plt.xlabel('Year' , fontdict={'fontname':'comic sans ms'})
plt.ylabel('Team',fontdict={'fontname':'Times New Roman'})
plt.show()
```

```python
# Plot a pie chart of trophy distribution
plt.figure(figsize=(8, 8))
plt.pie(trophy_counts, labels=trophy_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set3'))
plt.title('Percentage of IPL Trophies Won by Each Team (2008-2024)')
plt.show()
```
